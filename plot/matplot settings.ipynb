```python
x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x1, y1, label="Test Legend")
ax.set_xlabel('time [s]')
ax.set_ylabel('Damped oscillation [V]')
ax.set_title("Params Testing Plot")
ax.legend(loc="upper right")

plt.show()
```

# subplots tight layout

```python
import pandas as pd
df = pd.DataFrame({'x': [1, 2, 3, 4], 'y': [4, 3, 2, 1]})
df.plot(subplots=True, figsize=(8, 8), title='Subplots Tight Layout')
plt.gcf().tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```
